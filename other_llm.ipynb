````python
from transformers import AutoConfig, AutoTokenizer, AutoModel
config = AutoConfig.from_pretrained('zzxslp/RadBERT-RoBERTa-4m')
tokenizer = AutoTokenizer.from_pretrained('zzxslp/RadBERT-RoBERTa-4m')
model = AutoModel.from_pretrained('zzxslp/RadBERT-RoBERTa-4m', config=config)
text = """You are tasked with analyzing radiology reports to determine the presence or absence of Fracture of Leg. 
    Provide the output in JSON format, containing only the key 'Fracture of Leg' and its corresponding value of either 0 (absent) or 1 
    (present), based solely on the information in the report. The output should not include any additional text apart 
    from the JSON string. An exampleoutput format is: ```{'Fracture':0}```. Below is the radiology report to analyze: There is a hairline fracture in the Tibia just below the knee"""
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)

````
